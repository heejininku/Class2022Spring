```python
from transformers import Conversation 
#처음에는 transformer로부터 conversation이라는 library를 받아옴.
#이것은 다른 것과 다르게 중간과정이 조금 필요함.
conversation_1 = Conversation("Going to the movies tonight - any suggestions?")  #첫번째 문장을 변수
conversation_2 = Conversation("What's the last book you have read?")   #두번째 문장을 변수
pipe([conversation_1, conversation_2]) #리스트로 해서 pipeline에 넣는 것.
```

위의 결과값을 보면 user가 우리들이 되는 것이고, 각각 첫번째 문장, 두번째 문장에 따른 대답을 bot(챗봇)이 하고 있음

아래에서 위에 것을 그대로 받은 상태로 추가적으로 더 쓸 수도 있음.(bot의 발언 깔고 가는 것)

```python
conversation_1.add_user_input("Is it an action movie?")
conversation_2.add_user_input("What is the genre of this book?")
pipe([conversation_1, conversation_2])
```

위에 셀을 보면, 가장 맨 처음에 대화에서 덧붙여진 것을 확인할 수 있음.

```python
pipe = pipeline(task="feature-extraction") 
#feature를 가지고 sentence similarity 비교를 했었음. cosine similarity
#비교를 하기 위해서 feature를 뽑아줌_feature는 벡터 값(embedding 값)이 나오고 그것을 cosine similarity로 비교 가능
```

```python
pipe("What's the last book you have read?") 
#결과값을 보면 숫자들이 그냥 나옴_embedding 값
#이 숫자들을 가지고 비교하고자 하는 문장들의 유사성을 비교할 수 있음
```

```python
pipe = pipeline(task="zero-shot-classification") 
#zero-shot이라는 것은 그냥 막 질문해보는 것(원래 모델에게 학습을 시키는데 학습하지 않은 내용을)
```

```python
pipe("one day I will see the world", ['travel', 'cooking', 'dancing']) 
#이게 어떤 유형의 text인지 (3개 중에) 맞춰라_또는 확률을 알려달라
```

```python
pipe = pipeline(task="zero-shot-image-classification") 
#네이버,구글 렌즈로 이미지 보여주고 이것은 무슨 꽃인지 등을 알려주는 것과 비슷함
#단, 이것은 zero-shot이니까 한번도 학습하지 않은 것을 줌
```

```python
url = "http://images.cocodataset.org/val2017/000000039769.jpg"
pipe(url, candidate_labels=["a photo of a cat", "a photo of a dog" , "monkey"]) #이렇게 몇 개를 주면 확률을 알려줄 것
```
