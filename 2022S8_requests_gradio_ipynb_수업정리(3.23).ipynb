우선 API 가 무엇일까?
API=Application Programming Interface-> 워드나 한글 같이 컴퓨터에 깔려 있는 프로그램이면 모르겠지만, 다른 곳에 깔려 있는 서비스 등을 할 때 API라고 해서 개발자의 입장으로 연결(interface)해주는 것

예시) 미세먼지 앱, 맛집 앱, 관광 앱, 교통정보 앱 -> 이런 걸 어디서 가지고 오는가? 공공 데이터, 공공 api 라고 해서 언제든 쓸 수 있게 제공되고 있음. (제공하는 서버가 있음)

-> 제공해주는 것을 내 프로그램으로 만들 수 있게 (포함시키도록) 하는 것이 API임/ api 는 다양함 (카카오, 네이버도 있음. 음성인식, 자동번역 파파고 등을 내 앱이나 웹에 땡겨오고 싶을 때에 가져오는 게 api임) (무료,유료 api 모두 있음)

API 서비스를 땡겨올 수 있게 하는 패키지(라이브러리)가 request임.

지난 시간에 배운 것 중에 pd. series 사용해서 칼럼 아래에 들어가는 값들을 만들 것.7개의 칼럼을 만들 것 (아래에 pd, series 개수 세어보면 된다.)

이후에 dataframe 이라는 

```python
import pandas as pd
cityName = [item['cityName'] for item in items]  #list comprehension 적극 이용함. 25개의 아이템들 중에서 첫번째에 있는 리스트를 가지고 하나하나씩 for 이용해서 item(이름 상관 없음)으로 돌거다. 그런데 아이템들 중에서 city name의 dict 에 있는 value 값을 여기 city name에 넣어라.
pm10Value = [item['pm10Value'] for item in items] #아래에도 마찬가지로 주르륵 다 리스트로 잡힐 것임. 이 줄은 pm10value 값
pm25Value = [item['pm25Value'] for item in items]
so2Value = [item['so2Value'] for item in items]
coValue = [item['coValue'] for item in items]
o3Value = [item['o3Value'] for item in items]
no2Value = [item['no2Value'] for item in items]

cityName = pd.Series(cityName)
pm10Value = pd.Series(pm10Value)   
pm25Value = pd.Series(pm25Value)
so2Value = pd.Series(so2Value)
coValue = pd.Series(coValue)
o3Value = pd.Series(o3Value)
no2Value = pd.Series(no2Value)

df = pd.DataFrame({ 'cityName': cityName, 'pm10Value': pm10Value, 'pm25Value': pm25Value, 'so2Value': so2Value, 'coValue': coValue, 'o3Value': o3Value, 'no2Value': no2Value}) #칼럼명:칼럼내용 해서 짝 지어짐.
```

```python
def test(sidoName):
  URL = ('http://apis.data.go.kr/B552584/ArpltnStatsSvc/'
      'getCtprvnMesureSidoLIst?'
      'sidoName='+sidoName+
      '&searchCondition=hour'
      '&pageNo=1'
      '&numOfRows=100'
      '&returnType=json'
      '&serviceKey=miH%2BZXg85lQy4%2FkmhffvygXDIFiTwisriSLxtyECSzw6hxTVK7yI8QKIOc7mP7LEqCnhBGuj6iW1gZW65dum7A%3D%3D')
  
  response = requests.get(URL)
  r = response.json()
  items = r['response']['body']['items']

  cityName = [item['cityName'] for item in items]
  pm10Value = [item['pm10Value'] for item in items]
  pm25Value = [item['pm25Value'] for item in items]
  so2Value = [item['so2Value'] for item in items]
  coValue = [item['coValue'] for item in items]
  o3Value = [item['o3Value'] for item in items]
  no2Value = [item['no2Value'] for item in items]

  pd_cityName = pd.Series(cityName)
  pd_pm10Value = pd.Series(pm10Value)   
  pd_pm25Value = pd.Series(pm25Value)
  so2Value = pd.Series(so2Value)
  coValue = pd.Series(coValue)
  o3Value = pd.Series(o3Value)
  no2Value = pd.Series(no2Value)

  df = pd.DataFrame({ 'cityName': cityName, 'pm10Value': pm10Value, 'pm25Value': pm25Value, 'so2Value': so2Value, 'coValue': coValue, 'o3Value': o3Value, 'no2Value': no2Value})
  return df
```
