### Constructing a DataFrame using dict
df 만드는 방법 2번

```python
# Some input data - a small sample of the iris dataset
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}
#개수는 맞아야 함 (value 들 보면 위에도 5개, 아래도 5개). key 부분이 칼럼 이름이라고 생각하면 된다. 또 VALUE 쪽에는 리스트의 형태로 5개의 아이템이 들어감.
#하나의 KEY& VALUE가 한 칼럼이 되는 것임. DATA를 이미 dict로 만들고 아래의 입력부분에 dict (data)를 넣어주기
df = pd.DataFrame(data) 
#pandas에 들어있는 function인 DataFrame임. csv파일을 통해서 read해서 df를 불러올 수도 있지만, 지금처럼 df를 dict를 통해서 df 만들 수 있음.
df 
```

### Constructing a DataFrame using pd.Series

df 만드는 방법 3번.

한 칼럼에 해당하는 내용만을 pandas의 series함수로 만들 수 있음.

그리고 마지막에 pd.DataFrame 하고 

dict(칼럼명: 내용부분이름 아이템으로 이루어진)를 만들어주는 것임.

```python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df
```

이제는 우리가 이렇게 df 가 만들어진 것을
1. 필요한 특정정보만 indexing 을 해서 access 하는 방법을 배울 것.
2. 일부 내용 변경(수정)을 하는 방법을 배울 것.

### Changing column names

```python
df.columns
# df.columns 이렇게 해버리면 칼럼명만 등장함
```

```python
# Change the columns
#위에 보면 하나만 대문자 되어있어서 변경하고자 함.
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
#리스트로 변경해서 바꿔놓고 다시 assign하고 df 실행해보면 칼럼명이 바뀌어있음.
df
```

일부분만 가지고 오고 싶은 경우는 아래

### Accessing columns

```python
# Access one column 
df.loc[:, 'sepal_length']
#df.loc[:, 'sepal_length'] 하면 (loc은 location이라는 뜻) , 에 왼쪽에 있는 부분은 row임. , 오른쪽에 있는 부분은 column임.
#그냥 :이 오면 전부 다라는 뜻 (여기서는 행에 대해서 전부 다 라는 의미/ 칼럼은'sepal_length' 부분만 가져오게 되는 것. )
```

```python
# Access one column (short-hand)
# 더 간단히 하는 방법임 (위에 상황과 같은 것- 칼럼 명만 적으면 칼럼에 있는 모든 정보가 나옴)
df['sepal_length']
```

```python
# Access one column (shortest-hand)
#더더더 간단히 하는 방법임 (위에 상황과 같은 것- df. 하고 칼럼 명만 적으면 칼럼에 있는 모든 정보가 나옴)
df.sepal_length
```

두개의 칼럼도 가지고 올 수 있음.

```python
# Access two columns
df.loc[:, ['sepal_length', 'sepal_width']]
# 두개의 칼럼을 , 로 리스트로 연결해서 표현하면 가능함
```

```python
# Access two columns (short-hand)
#더 간단한 방법- 대괄호를 하고 리스트 적기
df[['sepal_length', 'sepal_width']]
```

이번에는 row 정보를 가지고 올 수도 있음.

### Accessing rows

```python
# Access one row #한 row 가져오기 
df.loc[3, :] # 몇번째 row와 모든 칼럼을 택하면 된다, (하나의 칼럼을 가지고 올 때와는 반대라고 생각하면 된다)
# 세로로 나오지만 row 정보임 (전체 표를 보면 알 수 있음)
```

```python
# Access two rows #두 개의 row를 가져오기
df.loc[[3, 0], :] # 몇번째 row를 원하는지 리스트로 적으면 된다.
```

```python
# Access a range of rows
df.loc[2:5, :]
# 이렇게 적게 된다면, 칼럼은 전부 다이고, 2,3,4 번째의 row만 가져오게 되는 것, (5번째는 제외됨)
```

```python
# Filter keeping only rows with `sepal_length` > 5
#유용하게 사용될 수 있음. -> sepal_length 5보다 큰 경우만 가지고 오는 것 (전체 표를 보면 0,1,3 row가 5보다 큼)
df[df.sepal_length > 5]
```

### Adding/Removing new columns

```python
df['new_column'] = 1
# df['new_column'] 이렇게 적는건 한 칼럼 access 하는 것임 이후에 1이 다 통으로 assign해버리는 것.
df
```

```python
df['new_column'] = [1, 2, 3, 4, 5]
# df['new_column'] 이렇게 적는건 한 칼럼 access 하는 것임 이후에 1,2,3,4,5를  assign해버리는 것.
df
```

```python
#한 칼럼을 빼고 싶을 때-> df = df.drop 하고, new column 을 찾아서 지우는 건데 이걸 칼럼에서 지울 수 있도록 # axis=1 이라고 적어준 것
#axis=0은 row 행 정보/ # axis=1 은 column 열 정보임
#그리고 꼭 시작에 df= 이라고 써야지만, df 업데이트가 가능함
df = df.drop('new_column', axis=1) # axis=1: column
df
```

이번에는 row를 더하거나 제거하는 방법임

### Adding/Removing new rows

새로운 row 만들기

```python
df.loc[5, :] = '' 
#row정보가  5로 써있음 (위에 표 보면 4까지 밖에 없기에 access 하면 에러이지만 뒤에 '' 이라는 아무것도 없는 blank 스트링을 assign해서 만듦
#칼럼은 모든 것에 대한 것이라는 얘기
#df 해보면 5번째 row는 스트링이 blank 인 걸로 채워져 있을 것임.
df
```

이번에는 row를 지우는 과정

```python
#한 row를 빼고 싶을 때-> df = df.drop 하고, 5라는 row를 찾아서 지우는 건데 이걸 row 중에서 지울 수 있도록 # axis=0 이라고 적어준 것
#axis=0은 row 행 정보/ # axis=1 은 column 열 정보임
#그리고 꼭 시작에 df= 이라고 써야지만, df 업데이트가 가능함
df = df.drop(5, axis=0) # axis=0: row
df
```
