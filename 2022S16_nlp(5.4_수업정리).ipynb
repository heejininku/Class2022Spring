### Normalization  

깨끗하게 하는 것임 (접미사,접두사 정리 처리 등등) 아래처럼 두가지 종류가 있음.

**Stemming** 어간 추출_ 대충의 패턴 규칙으로 어미를 잘라내는 것 (사전에 없는 어간 나올 수 있음)

**Lemmatization** 표제어(기본 사전형) 추출.


```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
[stemmer.stem(w) for w in words] #여기 words는 위에서 확인가능
#하나씩 w에 담아서 stemmer을 쓰는 것 (각각의 단어(토큰)에 대해서 쓸 수 있음)
```

```python
from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()
[stemmer.stem(w) for w in words]
```

```python
from nltk.stem import WordNetLemmatizer #표제어를 기반으로 원형으로 복구를 해내는 아이 (목적에 따라서는 위에 두개보다 정교)
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
[lemmatizer.lemmatize(w) for w in words]
```

### Stopword (구령어)

대명사, 조동사, 관사 처럼 너무 많이 사용되는 것은 우리가 관심이 없기에
너무 자주 나오는 단어는 따로 모아놓기도 함 리스트가 따로 있음 (100개정도되는것)

nltk에서 제공을 해줌.정의가 되어있음. 

```python
from nltk.corpus import stopwords  
nltk.download('stopwords') #정해진 단어의 리스트들을 받아왔고 그것을 기반으로 처리할 것
print(words) #stopwords가 있어도 그냥 내버려두는 아이
words = [w for w in words if not w in stopwords.words('english')] 
#words에 단어들이 있는데 이를 하나하나씩 받아서 for loof를 돌려봤다. 근데 그들 중 stopwords에 들어있지 않으면 그것을 취해라 (한마디로 stopwords 제거)  
print(words)
```

```python
nltk.Text(words).collocations()  
# ()안에 값을 써줄 수도 있지만 -> default(디폴트값이):(num=20, window_size=2) 총 20개의 collocations (두개의 단어가 연속된 것_이름이 많음) 
#window_size=2 (2개가 연어인 것.3을 하면 없어서 안나올 것)
```

```python
nltk.Text(words).concordance('Emma', 79, 10) 
#코퍼스 전체에서 엠마라는 단어가 등장하는, 앞뒤로의 context들을 같이 보여줌 (여기서는 10개 뽑기)(79는 하나 찝어서 캐릭터 수 세보면 79개일 것)
```

```python
nltk.Text(words).dispersion_plot(["Emma", "Knightley", "Frank", "Jane", "Harriet", "Robert"]) #교수님이 한번 사람이름을 넣어봄.
#dispersion_plot은 x축은 코퍼스 전체, 각각 엠마가 코퍼스의 어느부분에서 등장하는지 표시해주는 의미~
#여기서 누가 주인공이고 조연인지 등등을 알 수 있음
#사람 이름을 쓸 필요는 없고  다양한 단어들이 어떻게 분포되어 있는지를 배울 수 있는 것.
```

```python
# Distributional similarity: #분포적으로 유사한 것이 무엇인가
# find other words which appear in the same contexts as the specified word; #context는 양쪽 옆으로~라는 뜻_엠마랑 양쪽옆으로 나올만한 단어는?
# list most similar words first.
nltk.Text(words).similar("Emma") #엠마가 사람이니까 대명사가 많은 편
```

```python
# Find contexts where the specified words appear; list most frequent common contexts first.
nltk.Text(words).common_contexts(["Emma", "she"]) #세 개 넣어도 되는데 두 개 넣은 것. 이러한 단어가 공통으로 나타나는 양쪽 단어를 말해라
#그것도 역시 높은 것부터 순서대로 해라~ 
#의미는 but과 was 사이에, and와 could 사이에,,,, "Emma", "she"가 들어가기 쉽다~
```

규칙에 따라서 스마트하게 ~을 찾을 수 있는데 아래의 규칙을 참고하기

```python
'''       Basic Regular Expression Meta-Characters, Including Wildcards, Ranges and Closures

.	        Wildcard, matches any character
^abc	    Matches some pattern abc at the start of a string *시작부에서~
abc$	    Matches some pattern abc at the end of a string *끝에서~
[abc]	    Matches one of a set of characters
[^abc]    Matches anything but a set of characters
[A-Z0-9]	Matches one of a range of characters
ed|ing|s	Matches one of the specified strings (disjunction)
*	        Zero or more of previous item, e.g. a*, [a-z]* (also known as Kleene Closure)
+	        One or more of previous item, e.g. a+, [a-z]+
?	        Zero or one of the previous item (i.e. optional), e.g. a?, [a-z]?
{n}	      Exactly n repeats where n is a non-negative integer
{n,}	    At least n repeats
{,n}	    No more than n repeats
{m,n}	    At least m and no more than n repeats
a(b|c)+	  Parentheses that indicate the scope of the operators
(...)     Matches whatever regular expression is inside the parentheses
\d
Matches any decimal digit; this is equivalent to the class [0-9].
\D
Matches any non-digit character; this is equivalent to the class [^0-9].
\s
Matches any whitespace character; this is equivalent to the class [ \t\n\r\f\v].
\S
Matches any non-whitespace character; this is equivalent to the class [^ \t\n\r\f\v].
\w
Matches any alphanumeric character; this is equivalent to the class [a-zA-Z0-9_].
\W
Matches any non-alphanumeric character; this is equivalent to the class [^a-zA-Z0-9_].

'''
```

```python
engdict = nltk.corpus.words.words('en') #사전 가져오기
```

```python
engdict[10] #10번째 아이템 (스트링)-처럼 각각에 아이템에 사용할 것.
```

```python
re.search('ed$', 'educated') 
#이건 스트링 끝부분에서 ed를 찾으라는 말- 만약 $ 없이 ed만 적으면 앞쪽부터 찾음.
#'^ed' 이렇게 하면 스트링의 앞에서 나오는 애를 가져오는 것.
```

```python
result = [w for w in engdict if re.search('ed$', w)] #engdict라는 list가 있는데 list comprehension을 이용해서 results를 다시 만든다고 생각하기
#engdict에 있는 각각의 아이템을 하나하나씩 w로 받아서 그 개수만큼 for loof를 돈다 
#여기에 if적용됨_해당 부분을 찾아라~ 여기서 그냥 a를 찾아라 가 아니라 (ed|ing)$ 찾아라 이런식인데 이게 규칙 중 하나.(규칙이 적용됨)(위에 셀 참고)
#결국 engdict 중에서 ed로 끝나는 단어들만 다 모일 것
result = [w for w in engdict if re.search('^..j..t..$', w)] #...은 무슨의미?-어떤 것이든 와도 상관 없다 (.하나가 한 캐릭터)
result = [w for w in engdict if re.search('^[ghi][mno][jlk][def]$', w)]
#[ghi][mno][jlk][def]은? []은 이 안에 들어있는 것 중에 하나 [ghi]중에 하나 그다음은 [mno]중에 하나.. (결국 4개의 캐릭터가 있다는 의미)
result = [w for w in engdict if re.search('^[ah]+$', w)][:10] #[ah]+: a또는 h의 캐릭터가 한번 이상 나오는 것을 다 찾아라
print(result[:10]) #10개까지만 보여달라는 의미
```

```python
result = [w for w in wsj if re.search('(ed|ing)$', w)] #ed, ing로 끝나는 게 있기만 하면~
result = [w for w in wsj if re.search('^[0-9]+.[0-9]+\$', w)] #[0-9]: 숫자의미
#숫자 한번 이상 나오고 .이 나오고  숫자가 한번 이상 나오는 거 (사실 상 소수점)
#전화번호나 이메일도 형태가 잘 나타나기에 사용할 수 있는 영역
result = [w for w in wsj if re.search('^[A-Z]+\$$', w)] 
#\$는 끝이라는 의미가 아니라 진짜 달러모양(구분위해서 앞부분에 슬러시꺼꾸로기호가 앞에 들어감)
#어떤 대문자들이 하나 이상 나오고, 달러로 끝나는 거 찾아라
result = [w for w in wsj if re.search('^[0-9]{4}$', w)]
#{4}는 4번 나온다는 의미#결국 4자리 수 구하라는 의미
result = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)] 
#숫자 하나이상 나오고 -(데쉬 나오고)소문자가 3~5개 사이인것임 (ex.10-day)
result = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
#소문자 5개 이상-데시 나오고, 소문자가 2~3개 사이, 다시 데시나오고 소문자가 6개 이하
result= sorted(set(result))
print(result[:10])
```

```python
#pattern = '(?<=내가 찾고자 하는 것의 앞).(중간에 점을 찍으면 캐릭터 하나) (?=내가 찾고자 하는 것의 뒤)'
#결국 형태는 '(?<=  ).+(?=  )'
#가운데에 . 말고 .+이면 캐릭터가 한개 이상일 때임
pattern = '(?<=: ).+(?=[\.|\?|\!])' #왼쪽에 :이 있고, 오른쪽에는 . 또는 ? 또는 !
sent = re.findall(pattern, text) #이 패턴 만족시키는 예를 sent로 list화 해라
#길게 있는 글에서 findall이 유용_그 규칙에 해당하는 것을 다 찾아서 리스트화해줌
sent #len(sent)해보니까 6200개 정도 #여기는 각 리스트가 문장별로 몇천개 있는 것
```

```python
text = '\n'.join(sent) #리스트로 만든 것을 재결합시키기 
#줄바꿈('거꾸로슬러시n')으로
#만약에 한 칸 띄우고 싶으면 text= ' '.join(sent)
text
```
